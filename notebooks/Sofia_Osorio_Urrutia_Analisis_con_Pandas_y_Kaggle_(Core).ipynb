## Limpieza de Datos


```python
#Identifica y maneja los datos faltantes utilizando técnicas apropiadas
 #(relleno con valores estadísticos, interpolación, eliminación, etc.).
#df["Continent"].unique() #['Asia', 'Europe', 'Africa', 'North America', 'South America','Oceania', nan]
df[df["Continent"].isnull()] # hay dos filas con contienente Nan pero al parecer el continete está ubicado en la columna Country
#se reemplazan los valores de Continent por los de Country para el caso de Country = North America y "Oceania"
df.loc[df["Country"] == "North America", "Continent"] = "North America"
df.loc[df["Country"] == "Oceania", "Continent"] = "Oceania"
```

```python
#Corrige los tipos de datos si es necesario (por ejemplo, convertir cadenas a fechas).
#convertir Last_Updated_Date a tipo fecha datetime
df['Last_Updated_Date'] = pd.to_datetime(df['Last_Updated_Date'])
df.info()

```

```python
#Elimina duplicados si los hay.
df.duplicated().value_counts() #no hay duplicados
```

## 5. Transformación de Datos


```python
#Crea nuevas columnas basadas en operaciones con las columnas existentes
 #(por ejemplo, calcular ingresos a partir de ventas y precios).
#se crea la columna % de personas completamente vacunadas y se obtiene sobre la poblacion del pais
# se redondea a 2 decimales
df["%_de_people_fully_vaccinated"] = ((df["People_Fully_Vaccinated"] / df["Population"]) * 100).round(2)
#Normaliza o estandariza columnas si es necesario.
#como normalización se transforma el nombre de columnas a minusculas
df.columns = df.columns.str.lower()
```

```python
#Clasifica los datos en categorías relevantes.
#Crearé la columna >1000_deaths para visualizar los paises donde el total de muertes fue superior a las 1000

df[">1000_deaths"] = df["total_deaths"].apply(lambda x: "True" if x > 1000 else "False")
df.head(20)

#categorizar un pais segun su poblacion
def cat_population(x: int):
  if x <= 500000: return "micropais"
  elif x >=500000 and x <=4999999: return "pequeño"
  elif x >= 5000000 and x <= 49999999: return "mediano"
  elif x >= 50000000 and x <= 199999999: return "grande"
  elif x >= 200000000: return "muy grande"
#se categoriza tamaño de la poblacion del pais
df["categoria_poblacion"] = df["population"].apply(lambda x:cat_population(x) )
#confirmar si se aplico correctamente la funcion y se visualiza la nueva columna
df[df["country"]=="China"] #poblacion 1444216102	correspondería a gigante
df.head()
```

## 6. Análisis de Datos


```python
#Realiza agrupaciones de datos utilizando groupby para obtener insights específicos
 #(por ejemplo, ventas por producto, ventas por región, etc.).
 #Aplica funciones de agregación como sum, mean, count, min, max, std, y var.

# variabilidad de las muertes totales por país dentro de cada continente.
#oceania tiene una desviacion std relativamente baja (respecto a los otros continentes)
#lo que significa que la cantidad de muertes es similar en sus paises
df.groupby("continent")["total_deaths"].std()

#se visualiza la poblacion total y la poblacion completamnete vacunada de cada continente existente en el dataset
df.groupby("continent")[["population", "people_fully_vaccinated"]].sum()

#se visualiza la poblacion total y la poblacion completamnete vacunada agrupada por continente y pais
df.groupby(["continent", "country"])[["population", "people_fully_vaccinated"]].sum()

#se visualiza la poblacion total, la poblacion completamnete vacunada y el total de casos covid agrupada por continente y pais
df.groupby("continent")[["population", "people_fully_vaccinated", "total_cases"]].sum()

#se visualiza la poblacion total , la poblacion completamnete vacunada, el total de casos y total de muertes agrupada por continente y pais
df.groupby(["continent", "country"])[["population", "people_fully_vaccinated", "total_cases","total_deaths"]].sum()
```

```python
#Utiliza el método apply para realizar operaciones más complejas y personalizadas.
#porcentaje de muertes
#se obtiene % de muertes sobre total de casos de covid, en caso de que se cumpla la condicion de que total_cases sea >0. sino, se completa con 0.
#se visualiza agrupado con contiente y pais
df.groupby(["continent", "country"]).apply(lambda x: ( x["total_deaths"].sum() / x["total_cases"].sum()) * 100 if x["total_cases"].sum() > 0 else 0)
```
